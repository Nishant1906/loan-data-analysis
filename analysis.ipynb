```python
# write a function to cap outliers based on upper bound and lower bound IQR range

def clip_outliers(df):
    Q1 = df.quantile(0.25) 
    Q3 = df.quantile(0.75)
    IQR = Q3-Q1

    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    return df.clip(lower=lower_bound, upper=upper_bound)

def impute_missing_values(df, data_type):
    if data_type=='Numeric':
        median_value = df.median()
        return df.fillna(median_value)
    elif data_type == 'Categorical':
        mode_value = df.median()
        return df.fillna(mode_value)

```
